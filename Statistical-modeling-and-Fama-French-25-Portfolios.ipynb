# Statistical-modeling-and-Fama-French-25-Portfolios

```python
import numpy as np
import pandas as pd
import yfinance as yf
import pandas_datareader as pdr
import arch
import seaborn as sns
import warnings
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats
warnings.filterwarnings('ignore')
plt.style.use('seaborn')
```

## 1. (a) 

```python
ibm = yf.Ticker('IBM').history(start='1990-08-01', end='2019-08-01')
ibm.head()
```

```python
sp500 = yf.Ticker('^SP500TR').history(start='1990-08-01', end='2019-08-01')
sp500.head()
```

```python
ibm = ibm[['Close']]
sp500 = sp500[['Close']]
ibm['return'] = np.log(ibm['Close'] / ibm['Close'].shift(1))
sp500['return'] = np.log(sp500['Close'] / sp500['Close'].shift(1))
```

```python
ibm_models, sp500_models = {}, {}
update_freq = 10

# IBM
## ARCH(1)
ibm_models['ARCH(1)'] = arch.arch_model(ibm['return'].iloc[1:], p=1, q=0).fit(update_freq=update_freq)
## ARCH(2)
ibm_models['ARCH(2)'] = arch.arch_model(ibm['return'].iloc[1:], p=2, q=0).fit(update_freq=update_freq)
## GARCH(1,1)
ibm_models['GARCH(1,1)'] = arch.arch_model(ibm['return'].iloc[1:], p=1, q=1).fit(update_freq=update_freq)
## GARCH(2,1)
ibm_models['GARCH(2,1)'] = arch.arch_model(ibm['return'].iloc[1:], p=2, q=1).fit(update_freq=update_freq)
## TARCH(1,1,1)
ibm_models['TARCH(1,1,1)'] = arch.arch_model(ibm['return'].iloc[1:], p=1, o=1, q=1).fit(update_freq=update_freq)
## TARCH(1,1,1)
ibm_models['TARCH(2,2,1)'] = arch.arch_model(ibm['return'].iloc[1:], p=2, o=2, q=1).fit(update_freq=update_freq)

# S&P 500
## ARCH(1)
sp500_models['ARCH(1)'] = arch.arch_model(sp500['return'].iloc[1:], p=1, q=0).fit(update_freq=update_freq)
## ARCH(2)
sp500_models['ARCH(2)'] = arch.arch_model(sp500['return'].iloc[1:], p=2, q=0).fit(update_freq=update_freq)
## GARCH(1,1)
sp500_models['GARCH(1,1)'] = arch.arch_model(sp500['return'].iloc[1:], p=1, q=1).fit(update_freq=update_freq)
## GARCH(2,1)
sp500_models['GARCH(2,1)'] = arch.arch_model(sp500['return'].iloc[1:], p=2, q=1).fit(update_freq=update_freq)
## TARCH(1,1,1)
sp500_models['TARCH(1,1,1)'] = arch.arch_model(sp500['return'].iloc[1:], p=1, o=1, q=1).fit(update_freq=update_freq)
## TARCH(1,1,1)
sp500_models['TARCH(2,2,1)'] = arch.arch_model(sp500['return'].iloc[1:], p=2, o=2, q=1).fit(update_freq=update_freq)
```

#### IBM Model Results

```python
print('IBM\n', ibm_models)
```

#### S&P 500 Index Model Results

```python
print('S&P 500 Index\n', sp500_models)
```

### (b) 

```python
ibm_bic = pd.DataFrame(dict(zip(list(ibm_models.keys()), list([ibm_models[model].bic for model in list(ibm_models.keys())]))), index=['IBM_BIC']).T
sp500_bic = pd.DataFrame(dict(zip(list(sp500_models.keys()), list([sp500_models[model].bic for model in list(sp500_models.keys())]))), index=['S&P500_BIC']).T
bic = pd.merge(ibm_bic, sp500_bic, left_index=True, right_index=True)
bic
```


- For IBM, we should choose TARCH(1,1,1);
- For S&P 500 Index, we should choose TARCH(1,1,1).

### (c)

```python
ibm_VaR = 100 * stats.norm.ppf(.95) * np.sqrt(ibm_models['TARCH(1,1,1)'].forecast(start='1990-08-01', align='target').variance)
ibm_VaR.columns = ['IBM_VaR']
sp500_VaR = 100 * stats.norm.ppf(.95) * np.sqrt(sp500_models['TARCH(1,1,1)'].forecast(start='1990-08-01', align='target').variance)
sp500_VaR.columns = ['S&P500_VaR']
VaR = pd.merge(ibm_VaR, sp500_VaR, left_index=True, right_index=True)

fig, ax = plt.subplots(figsize=(18, 6))
VaR.plot(ax=ax, fontsize=15)
ax.set_ylabel('VaR (in million dollars)', fontsize=15)
ax.set_xlabel('')
ax.set_title('VaR of IBM and S&P 500 Index', fontsize=25)
ax.legend(loc=0, fontsize=15)
```

## 2  (a) 

```python
fm = pdr.get_data_famafrench('25_Portfolios_5x5', start='1963-07', end='2019-09')[0]
fm.head()
```

### (b)

```python
factors = pdr.get_data_famafrench('F-F_Research_Data_Factors', start='1963-07', end='2019-09')[0]
factors.head()
```

### (c) 

```python
results = pd.DataFrame(columns=fm.columns, index=['alpha', 'alpha_t', 'Mkt-Rf', 'SMB', 'HML'])
fm_models = {}
for column in fm.columns:
    model = sm.OLS(fm[column] - factors['RF'], sm.add_constant(factors.iloc[:, :3])).fit()
    _res = model.params.tolist()
    _res.insert(1, model.tvalues[0])
    results[column] = _res
    fm_models[column] = model
results
```

### (d) 

```python
fig, axes = plt.subplots(5, 5, figsize=(80, 40))
for i, column in enumerate(fm.columns.tolist()):
    ax = axes[i % 5][i // 5]
    portfolio = pd.DataFrame(fm[column] - factors['RF'], columns=['Realized'])
    portfolio['Predicted'] = fm_models[column].predict()
    portfolio.plot(ax=ax, fontsize=10)
    ax.set_title('Risk Premium of %s' % column, fontsize=20)
    ax.set_xlabel('')
    ax.legend(loc=0, fontsize=15)
```
